### Import the needed

```python
import fastbook
from fastai.vision.all import * 
from fastai.vision.widgets import *
```

### create a Path Object & check for the saved Model File in .pkl format

```python
path = Path()
path.ls(file_exts='.pkl')
```

### For inference, load the saved model file

```python
learn_inf = load_learner(path/"export.pkl")
```

### Passing an Image for prediction

```python
learn_inf.predict('grizzlybear.jpeg')
```

### Accesing the categories of dataloader sorted

```python
learn_inf.dls.vocab
```

### Creating a Fileupload widget

```python
btn_upload = widgets.FileUpload()
btn_upload
```

### Create an image object from the uploaded image

```python
img = PILImage.create(btn_upload.data[0])
```

### Output Widget to display the image

```python
out_pl = widgets.Output()
out_pl.clear_output()
with out_pl: display(img.to_thumb(128,128))
out_pl
```

### Predicting the image

```python
pred,pred_idx,probs = learn_inf.predict(img)
```

### Using a label to display the output

```python
lbl_pred = widgets.Label()
lbl_pred.value = f'Prediction: {pred}; Probability: {probs[pred_idx]:.04f}'
lbl_pred
```

### Create a classify button to use for classification

```python
btn_run = widgets.Button(description='Classify')
btn_run
```

### Function to execute when classify button is called



```python
def on_click_classify(change):
    img = PILImage.create(btn_upload.data[-1])
    out_pl.clear_output()
    with out_pl: display(img.to_thumb(128,128))
    pred,pred_idx,probs = learn_inf.predict(img)
    lbl_pred.value = f'Prediction: {pred}; Probability: {probs[pred_idx]:.04f}'

btn_run.on_click(on_click_classify)
```

```python
btn_upload = widgets.FileUpload()
```

### A vertical box (VBox) to display nice GUI

```python
display(VBox([widgets.Label('Select your bear!'), btn_upload, out_pl, lbl_pred]))
```
